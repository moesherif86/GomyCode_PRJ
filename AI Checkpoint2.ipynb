```python
# the below function generates a heat map using seaborn
# and matplotlib libraries, the function removes the irrelevant columns
# that won't affect the study like the names of the passengers and only
# keep those columns that we can correlate creating a 2 dimensional matrix.
# in the heat map we are guided by a color bar on the right.
# as per the below code result it shows that there is important correlation
# between Parch & SibSp factors that highly affect survivability  
# while passenger ID was the least factor to affect survivability.

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
    
plot_correlation_map(df)
```
